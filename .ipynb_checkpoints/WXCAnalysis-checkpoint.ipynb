```python
data_set = {}
for item in range(len(df)):
    team = df.iloc[item]['Team']
    place = df.iloc[item]['New_Place']
    if team not in data_set:
        data_set[team] = [place]
        #print('yes',team,place)
    else:
        # this adds all 7 of the team members places and gives a one number total
        #data_set[team] += place
        # this puts all 7 team members places in a list
        data_set[team].append(place)
        #print('pass')

s = [(k, data_set[k]) for k in sorted(data_set, key=data_set.get, reverse=False)]

# use the below code to print out the team name, team score for runners 1 - 5, and then each runners place
# copy and paste into text file and reimport and sort as a df
#for k, v in s:
#    team_total = 0
#    for i in range(0,5):
#        team_total += v[i]
#    print('{};{};{}'.format(k,team_total,v)) 
#
#    #prints the team team and whatever the value is set to above
#    #print(k,v)

#for k, v in data_set.items():
#    print(k, v)    
    
#print(df.iloc[item]['Team'], df.iloc[item]['New_Place'])

with open('FinalData.csv','w') as file:
    for line in open('FinalData.txt','r'):
        row = line.strip().split(';') 
        writer = csv.writer(file)
        writer.writerow(row)

# import data and label columns
final_df = pd.read_csv('FinalData.csv', names=['Team', 'Team Score', 'Individuals Place'])
#df = df.sort_values(by=["Minutes","Seconds"])
final_df = final_df.sort_values(by=['Team Score'])
final_df
```
